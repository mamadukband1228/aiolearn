```python
import random

# تعریف لیست بازی با مخفف‌ها
bazi = ["a", "b", "c"]  # a برای sang، b برای kaqaz، c برای qeychi

# امتیازها
you_point = 0
computer_point = 0

# شرط ادامه بازی
shart_bazi = True

print("you        com")

while shart_bazi:
    you = input("Entekhab kon (a: sang, b: kaqaz, c: qeychi): ")
    if you == "0":
        shart_bazi = False

    rand = random.randint(0, 2)
    computer = bazi[rand]

    # بررسی برنده
    if you == 'a' and computer == 'b':
        computer_point += 1
    elif you == 'c' and computer == 'b':
        you_point += 1
    elif you == 'a' and computer == 'c':
        you_point += 1
    elif you == 'b' and computer == 'c':
        computer_point += 1
    elif you == 'c' and computer == 'a':
        computer_point += 1
    elif you == 'b' and computer == 'a':
        you_point += 1

    # بررسی شرط پایان بازی
    if you_point >= 3 or computer_point >= 3:
        shart_bazi = False

    # چاپ امتیازها
    print(you_point, end="          ")
    print(computer_point)

# اعلام نتیجه نهایی
if you_point > computer_point:
    print("\nYOU WIN!")
else:
    print("\nYOU LOSE :(")

print("YOU: ", you_point)
print("COM: ", computer_point)
```
